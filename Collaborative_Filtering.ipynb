#### RMSE

```python
def get_predicted_rating(ind_tuple):
    return round(y_hat[ind_tuple[0]-1, ind_tuple[1]-1].item(), 1)

def get_train_rmse():
    temp_train_user_movie_rating_df['pred_rating'] = list(map(get_predicted_rating, temp_train_user_movie_rating_df.ind))
    temp_train_user_movie_rating_df['diff'] = temp_train_user_movie_rating_df.rating-temp_train_user_movie_rating_df.pred_rating
    temp_train_user_movie_rating_df['diff2'] = temp_train_user_movie_rating_df['diff']**2
    return temp_train_user_movie_rating_df['diff2'].mean()

def get_val_rmse():
    temp_val_user_movie_rating_df['pred_rating'] = list(map(get_predicted_rating, temp_val_user_movie_rating_df.ind))
    temp_val_user_movie_rating_df['diff'] = temp_val_user_movie_rating_df.rating-temp_val_user_movie_rating_df.pred_rating
    temp_val_user_movie_rating_df['diff2'] = temp_val_user_movie_rating_df['diff']**2
    return temp_val_user_movie_rating_df['diff2'].mean()
```

## Training

#### Device

```python
device = 'cuda' if torch.cuda.is_available() else 'cpu'
```

#### Number of users, movies and features

```python
n_users, n_movies = train_rating.shape
n_features = 23
n_users, n_movies, n_features
```

#### Random Initialization

```python
torch.manual_seed(0)
users_emb = torch.randn((n_features,n_users), requires_grad=True, dtype=torch.float, device=device)
movies_emb = torch.randn((n_features,n_movies), requires_grad=True, dtype=torch.float, device=device)
```

#### Optimizer

```python
lr = 0.6

optimizer = optim.SGD([users_emb, movies_emb], lr=lr, momentum=0.9)
optimizer
```

#### Train and evaluate

```python
epochs = 1000

for epoch in range(epochs):
    
    y_hat = torch.mm(users_emb.T, movies_emb) * train_r
    loss = (((train_rating - y_hat)**2).sum()) / n_train_ratings
    loss.backward()
    optimizer.step()
    optimizer.zero_grad()
    if epoch % (epochs//10)==0:
        with torch.no_grad():
            y_hat = (torch.mm(users_emb.T, movies_emb))
            y_hat[y_hat<0]=0
            y_hat[y_hat>5]=5
            train_loss = get_train_rmse()
            val_loss = get_val_rmse()
        print(epoch, loss.item(), train_loss, val_loss)
```

## Precision@k and Recall@k

### Load Embeddings

```python
users_emb = torch.load('embeddings/users_emb.pt')
movies_emb = torch.load('embeddings/movies_emb.pt')
```

####  Training rating table

```python
train_rating_df = pd.read_csv('data/train_rating.csv')
train_rating_df.head(2)
```

#### Predictive ratings

#### Predictive rating table

```python
pred_rating_df = pd.DataFrame(pred_rating, columns=train_rating_df.drop(['user_id'], axis=1).columns, 
                              index=train_rating_df.user_id)
pred_rating_df.reset_index(inplace=True)
pred_rating_df.head(2)
```

```python
pred_user_movie_rating_df = pred_rating_df.melt(id_vars=['user_id'], value_vars=pred_rating_df.drop(['user_id'],
                                axis=1).columns, var_name='movie_id', value_name='rating')
pred_user_movie_rating_df.head(3)
```

#### All user-movie-ratings data

```python
all_user_movie_rating_df = train_user_movie_rating_df.copy()
all_user_movie_rating_df['pred_rating'] = pred_user_movie_rating_df.rating
all_user_movie_rating_df.head(3)
```

```python
df = all_user_movie_rating_df.copy()
```

```python
def get_precision_recall_at_k(df, user_id, k=10, threshold=4.5):
    
    specific_user_movie_rating_df = df.loc[(df.user_id==user_id)&(df.rating!=0)]
    specific_user_movie_rating_df.sort_values(by='pred_rating', axis=0, ascending=False, inplace=True)
    specific_user_movie_rating_df = specific_user_movie_rating_df[0:k]
    specific_user_movie_rating_df.loc[specific_user_movie_rating_df.rating>=threshold, 'relevant'] = 1
    specific_user_movie_rating_df.loc[specific_user_movie_rating_df.pred_rating>=threshold, 'recommend'] = 1
    specific_user_movie_rating_df.fillna(0, inplace=True)
    
    relevant = len(specific_user_movie_rating_df.relevant.loc[specific_user_movie_rating_df.relevant==1])
    recommend = len(specific_user_movie_rating_df.recommend.loc[specific_user_movie_rating_df.recommend==1])
    relevant_recommend = len(specific_user_movie_rating_df[['relevant', 'recommend']].loc[(
        specific_user_movie_rating_df.relevant==1)&(specific_user_movie_rating_df.recommend==1)])
    
    return(relevant_recommend/(relevant+1e-10), relevant_recommend/(recommend+1e-10))
```

```python
def get_metrics(df, k=10, threshold=4.5):
    
    precision_list = []
    recall_list = []
    users_list = list(all_user_movie_rating_df.user_id.unique())
    for user_id in users_list:
        precision, recall = get_precision_recall_at_k(df, user_id, k, threshold)
        precision_list.append(precision)
        recall_list.append(recall)
    return pd.DataFrame({'user':users_list, 'precision@'+str(k):precision_list, 'recall@'+str(k):recall_list})
```

#### Precison@k and Recall@k

k : number of movies to be recommended

rating >= threshold is relevant

predicted rating >= threshold is recommendable

```python
k = 5
threshold = 4

metric_df = get_metrics(all_user_movie_rating_df, k=k, threshold=threshold)
print('Mean P@k : ', metric_df['precision@'+str(k)].mean(), '| Mean R@k : ', metric_df['recall@'+str(k)].mean())
metric_df.head(2)
```

```python
k = 10
threshold = 4

metric_df = get_metrics(all_user_movie_rating_df, k=k, threshold=threshold)
print('Mean P@k : ', metric_df['precision@'+str(k)].mean(), '| Mean R@k : ', metric_df['recall@'+str(k)].mean())
metric_df.head(2)
```
