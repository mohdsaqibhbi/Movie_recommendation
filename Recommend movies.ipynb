```python
import pandas as pd
import numpy as np
import torch
import warnings
warnings.filterwarnings("ignore")

pd.options.display.max_columns = 200
pd.options.display.max_rows = 200
```

## Data Preparation

###  Training rating table

```python
train_rating_df = pd.read_csv('data/train_rating.csv')
train_rating_df.head(2)
```

```python
train_rating_df.replace(0, np.nan, inplace=True)
train_rating_df.head(2)
```

### Load Embeddings

```python
users_emb = torch.load('embeddings/users_emb.pt')
movies_emb = torch.load('embeddings/movies_emb.pt')
```

### Predictive ratings

```python
y_hat = (torch.mm(users_emb.T, movies_emb))
y_hat[y_hat<1]=1 # replacing with minimum rating
y_hat[y_hat>5]=5 # replacing with maximum rating

pred_rating = y_hat.numpy().round(1)
```

### Predictive rating table

```python
pred_rating_df = pd.DataFrame(pred_rating, columns=train_rating_df.drop(['user_id'], axis=1).columns, 
                              index=train_rating_df.user_id)
pred_rating_df.head(2)
```

```python
pred_rating_df.reset_index(inplace=True)
pred_rating_df.head(2)
```

### Whole rating table

```python
whole_rating_df = train_rating_df.combine_first(pred_rating_df)
whole_rating_df.head(3)
```

#### Write to csv

```python
whole_rating_df.to_csv('data/whole_rating.csv', index=False)
```

### Validation user-movie-ratings data

```python
val_user_movie_rating_df = pd.read_csv('data/val_user_movie_rating.csv')
val_user_movie_rating_df.head(2)
```

### Whole user-movie-ratings data

```python
whole_user_movie_rating_df = whole_rating_df.melt(id_vars=['user_id'], value_vars=whole_rating_df.drop(['user_id'],
                                axis=1).columns, var_name='movie_id', value_name='rating')
whole_user_movie_rating_df.head(2)
```

#### Mention watched/not watched movies by user

```python
for user_id, movie_id in zip(val_user_movie_rating_df.user_id, val_user_movie_rating_df.movie_id):
    whole_user_movie_rating_df.loc[(whole_user_movie_rating_df.user_id == user_id)&(whole_user_movie_rating_df.movie_id == str(movie_id)), 'status'] = "not watched"
    
whole_user_movie_rating_df.head(2)
```

```python
whole_user_movie_rating_df.fillna("watched", inplace=True)
```

```python
whole_user_movie_rating_df.head(2)
```

```python
whole_user_movie_rating_df = whole_user_movie_rating_df.astype({'movie_id': 'int64'})
whole_user_movie_rating_df.dtypes
```

#### Write to csv

```python
whole_user_movie_rating_df.to_csv('data/whole_user_movie_rating.csv', index=False)
```

### Training user-movie-ratings data

```python
train_user_movie_rating_df = pd.read_csv('data/train_user_movie_rating.csv')
train_user_movie_rating_df.head(2)
```

### Movies data

```python
movies_df = pd.read_csv('data/movies.csv')
movies_df.head(2)
```

#### Movie Indexing : movie_id to movie genre

```python
movies_genre_df = movies_df.drop(['movie_id'], axis=1)

genre_list = ["|".join(list(movies_genre_df.columns[(movies_genre_df == 1).iloc[movie_id-1]])) 
                     for movie_id in movies_df.movie_id]

id_to_genre_dict = dict(zip(movies_df.movie_id, genre_list))
```

#### Movie indexing : movie_id to movie_title

```python
id_to_title_dict = dict(zip(movies_df.movie_id, movies_df.movie_title))
```

#### Get user's taste

```python
def get_user_taste(train_user_movie_rating_df, id_to_title_dict, id_to_genre_dict, user_id):
    taste_df = train_user_movie_rating_df.loc[train_user_movie_rating_df.user_id==user_id].sort_values(
    by=['rating'], ascending=False)
    taste_df["movie_title"] = [id_to_title_dict[movie_id] for movie_id in taste_df.movie_id]
    taste_df["genre"] = [id_to_genre_dict[movie_id] for movie_id in taste_df.movie_id]
    
    return taste_df
```

```python
user_id = 94
get_user_taste(train_user_movie_rating_df, id_to_title_dict, id_to_genre_dict, user_id).head(20)
```

#### Recommend to user

```python
def recommend_to_user(whole_user_movie_rating_df, id_to_title_dict, id_to_genre_dict, user_id, k=5):
    recommend_df = whole_user_movie_rating_df.loc[(whole_user_movie_rating_df.user_id==user_id)&(
        whole_user_movie_rating_df.status=="not watched")].sort_values(by=['rating'], ascending=False)[:k]
    recommend_df["movie_title"] = [id_to_title_dict[movie_id] for movie_id in recommend_df.movie_id]
    recommend_df["genre"] = [id_to_genre_dict[movie_id] for movie_id in recommend_df.movie_id]
    recommend_df.rename(columns={'rating': 'pred_rating'}, inplace=True)

    return recommend_df
```

```python
k = 10
user_id = 94
recommend_to_user(whole_user_movie_rating_df, id_to_title_dict, id_to_genre_dict, user_id, k)
```
